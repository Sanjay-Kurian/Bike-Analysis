```python
staff = pd.read_sql("SELECT * FROM dbo.staffs", conn)
print(staff)
```

```python
sales_amount = pd.read_sql("SELECT year(o.order_date) as year_of_sale, month(o.order_date) as month_of_sale, SUM((oi.quantity * oi.list_price) - oi.discount) as total_sales_amount "
                            "FROM dbo.orders o " 
                            "JOIN dbo.order_items oi ON o.order_id = oi.order_id " 
                            "JOIN dbo.stores s ON s.store_id = o.store_id " 
                            "WHERE o.store_id = 1 "
                            "GROUP BY year(o.order_date), month(o.order_date) "
                            "ORDER BY year(o.order_date), month(o.order_date) ASC ",conn)
sales_amount
```

```python
rank = pd.read_sql("SELECT s.first_name, s.last_name, COUNT(o.order_id) as [number of orders], "
                   "RANK() OVER (ORDER BY COUNT(o.order_id) DESC) rank "
                   "FROM staffs s "
                   "JOIN orders o "
                   "ON s.staff_id = o.staff_id "
                   "GROUP BY s.first_name, s.last_name ",conn)
rank
```

```python
shipped_late = pd.read_sql("SELECT *, CASE WHEN shipped_date > required_date "
                           "THEN 1 "
                           "ELSE 0 "
                           "END AS 'shipped_late' "
                           "FROM orders ",conn)
shipped_late
```

```python
datatype = pd.read_sql("SELECT order_id, CAST(order_id AS float) AS order_id_float "
                       "FROM orders",conn)
datatype
```

```python
recentorders = pd.read_sql("SELECT o1.customer_id, MAX(o1.order_date) AS [recent order], "
                           "MAX(o2.order_date) AS [second most recent order] "
                           "FROM orders o1 "
                           "JOIN orders o2 ON o1.customer_id = o2.customer_id "
                           "AND o1.order_date > o2.order_date "
                           "GROUP BY o1.customer_id", conn)
recentorders
```

```python
query = """
    WITH employee_hierarchy AS (
                 SELECT staff_id, first_name + ' ' + last_name AS full_name, manager_id, 1 AS level
                 FROM staffs s1 
                 WHERE manager_id IS NULL 
                 UNION ALL
                 SELECT s2.staff_id, s2.first_name + ' ' + s2.last_name AS full_name, s2.manager_id, level + 1 AS lvl
                 FROM employee_hierarchy
                 JOIN staffs s2
                 ON s2.manager_id = employee_hierarchy.staff_id)
    SELECT * 
    FROM employee_hierarchy;
"""

eh = pd.read_sql(query,conn)
eh
```

```python
query = """
WITH stats AS (SELECT o.customer_id, SUM(oi.quantity * oi.list_price * (1 - oi.discount)) as total_cost,
COUNT(DISTINCT o.order_id) AS total_orders,
DATEDIFF(DAY, '2018-12-29', MAX(o.order_date)) AS days_since_last_purchase
FROM order_items oi
JOIN orders o
ON o.order_id = oi.order_id
GROUP BY o.customer_id
)
SELECT customer_id, CASE WHEN total_orders > 1 THEN 'repeat buyer'
                        ELSE 'one time buyer'
                        END AS purchase_frequency,
                        CASE WHEN days_since_last_purchase < 90 THEN 'recent buyer'
                        ELSE 'not recent buyer'
                        END AS purchase_recency,
                        CASE WHEN total_cost/MAX(total_cost) OVER() >= .65 THEN 'big spender'
                        WHEN total_cost/MAX(total_cost) OVER() <= .30 THEN 'low spender'
                        ELSE 'average spender' 
                        END AS buying_power
FROM stats;
"""
customers = pd.read_sql(query, conn)
customers
```

```python
query = """
SELECT c.first_name + ' ' + c.last_name AS full_name, COUNT(o.order_id) AS number_of_purchases,
RANK() OVER(ORDER BY COUNT(o.order_id) DESC) AS rank
FROM orders o
JOIN customers c
ON o.customer_id = c.customer_id
GROUP BY c.first_name, c.last_name;
"""
rank = pd.read_sql(query, conn)
rank
```

```python
query = """
SELECT p1.product_name, p2.product_name, COUNT(*) AS frequently_bought 
FROM order_items oi1
JOIN order_items oi2
ON oi1.order_id = oi2.order_id
AND oi1.product_id > oi2.product_id
JOIN products p1
ON oi1.product_id = p1.product_id
JOIN products p2
ON oi2.product_id = p2.product_id
GROUP BY p1.product_name, p2.product_name
ORDER BY frequently_bought DESC;
"""
basket = pd.read_sql(query, conn)
basket
```
